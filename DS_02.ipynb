```python
import numpy as	np
df7['DBMS']=df7['DBMS'].fillna(df7['DBMS'].mean())
Q1=	np.percentile(df7['DBMS'],	25,	interpolation='midpoint')
Q3=np.percentile(df7['DBMS'],75,interpolation='midpoint')

IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df7[(df7['DBMS']<lower_bound)|(df7['DBMS']>upper_bound)]
print('outliers')
print(outliers)
print('lowerbound',lower_bound)
print('upperbound',upper_bound)
```

```python
df7_no_outliers=df7[(df7['DBMS']>=lower_bound)&(df7['DBMS']<=upper_bound)]
print("Data without outliers:")
print('DBMS')
print(df7_no_outliers['DBMS'])
```

```python
import numpy as	np
df7['DS']=df7['DS'].fillna(df7['DS'].median())	#we	can	not	calculate	outlier	with	null	values
Q1=np.percentile(df7['DS'],25,interpolation='midpoint')
Q3=np.percentile(df7['DS'],75,interpolation='midpoint')
#IQR
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers	=	df7[(df7['DS']	<	lower_bound)	|	(df7['DS']	>	upper_bound)]
print('outliers')
print(outliers)
print('lowerbound',lower_bound)
print('upperbound',upper_bound)
```

```python
df7_no_outliers=df7[(df7['DS']>=lower_bound)&(df7['DS']<=upper_bound)]
print("Data	without	outliers:")
print('DS')
print(df7_no_outliers['DS'])
```

```python
import	pandas	as	pd
 #also	we	can	calculate	q1	and	q3	as	given	below
Q1	=	df7['OS'].quantile(0.25)
Q3	=	df7['OS'].quantile(0.75)
 #	Calculate	IQR
IQR	=	Q3	-	Q1
 #	Define	lower	and	upper	bounds
lower_bound	=	Q1	-	1.5	*	IQR
upper_bound	=	Q3	+	1.5	*	IQR
 #	Identify	outliers
outliers	=	df7[(df7['OS']	<	lower_bound)	|	(df7['OS']	>	upper_bound)]
print("Outliers:")
print(outliers)
print('lowerbound',lower_bound)
print('upperbound',upper_bound)
```

```python
df7_no_outliers	=	df7[(df7['OS']	>=	lower_bound)	&	(df7['OS']	<=	upper_bound)]
print("Data	without	outliers:")
print('OS')
print(df7_no_outliers['OS'])
```

```python
from	sklearn.datasets	import	load_iris
import	pandas	as	pd
iris	=	load_iris()#	Load	the	dataset
df	=	pd.DataFrame(iris.data,	columns=iris.feature_names)
print(df.head())
```

```python
from sklearn.preprocessing	import	MinMaxScaler
scaler=MinMaxScaler()
normalized_data=scaler.fit_transform(df)		
normalized_df=pd.DataFrame(normalized_data,	columns=iris.feature_names)
print("Normalized	Data:")
print(normalized_df.head())
```

```python
from	sklearn.preprocessing	import	StandardScaler
scaler	=	StandardScaler()
standardized_data=scaler.fit_transform(df)
standardized_df=pd.DataFrame(standardized_data,columns=iris.feature_names)
print("Standardized Data:")
print(standardized_df.head())
```
