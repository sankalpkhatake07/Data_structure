```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df1=pd.read_csv('demo.csv')
```

```python
df1
```

```python
df1.dropna(how='all',inplace=True)
df1.replace(to_replace=np.nan,value=10,inplace=True)
df1['DS'].replace(to_replace=393,value=66,inplace=True)
df1['DS'].replace(to_replace=10,value=56,inplace=True)
df1
```

```python
sns.boxplot(df1['DS'])
```

```python
#To calculate Outliers we use Interquartile Range
```

```python
Q1=np.percentile(df1['DS'],25,interpolation='midpoint')
Q3=np.percentile(df1['DS'],75,interpolation='midpoint')
IQR=Q3-Q1
```

```python
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
upper = df1['DS']>=(Q3+1.5*IQR)
print('Upper:',upper)
print(np.where(upper))
lower = df1['DS']<=(Q1-1.5*IQR)
print('Lower:',lower)
print(np.where(lower))
```

```python
UB = Q3 + 1.5 * IQR
LB = Q1 - 1.5 * IQR
outliers_above = df1['DS'] >= UB
outliers_below = df1['DS'] <= LB
print("Upper Bound:", UB)
print("Lower Bound:", LB)
print("Outliers above UB:", df1[outliers_above])
print("Outliers below LB:", df1[outliers_below])
```

```python
print("Old Shape: ",df1.shape)
upper=np.where(df1['DS']>=(Q3+1.5*IQR))
lower=np.where(df1['DS']<=(Q1-1.5*IQR))
df1.drop(upper[0],inplace=True)
df1.drop(lower[0],inplace=True)
print('New Shape: ',df1.shape)
```
