```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
plt.hist(df['sepal_length'])
plt.xlabel("sepal_length")
plt.title("Histogram for Sepal length")
```

```python
plt.hist(df['sepal_width'])
plt.xlabel("sepal_width")
plt.title("Histogram for Sepal width")
```

```python
plt.hist(df['petal_length'] ,bins=20,color='purple')
plt.xlabel("petal_length")
plt.title("Histogram for petal length")
```

```python
plt.hist(df['petal_width'],bins=20,color='pink')
plt.xlabel("petal_width")
plt.title("Histogram for petal width")
```

```python
plt.boxplot(df["sepal_length"])
plt.xlabel("sepal_length")
plt.title("Boxplot for sepal length")
```

```python
plt.boxplot(df["sepal_width"])
plt.xlabel("sepal_width")
plt.title("Boxplot for sepal width")
```

```python
plt.boxplot(df["petal_length"])
plt.xlabel("petal_length")
plt.title("Boxplot for petal length")
```

```python
plt.boxplot(df["petal_width"])
plt.xlabel("petal_width")
plt.title("Boxplot for petal width")
```
